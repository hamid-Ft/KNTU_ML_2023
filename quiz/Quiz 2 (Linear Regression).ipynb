```python
# Do it.
print(df.isnull().values.any())

df = df.dropna()
# scale features if you want . . . (uncomment following line)
#df = (df - df.mean()) / df.std()

print(df.head(10))
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font size=4>
<ol dir=rtl start="4">
<li>
از میان <b>5</b> ویژگی <code>Rooms</code> (تعداد اتاق)، <code>Bathroom</code> (تعداد حمام)، <code>Landsize</code> (اندازه زمین)، <code>Lattitude</code> (عرض جغرافیایی) و <code>Longtitude</code> (طول جغرافیایی) در این مجموعه‌داده، به ترتیب کدام ویژگی‌ها، تأثیر بیشتری بر قیمت خانه دارند؟ با رسم نمودار، به این سوال پاسخ دهید.
</li>
</ol>
</font>
</p>

```python
fig, axs = plt.subplots(2, 3, figsize=(30,12))

features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
titles = ['Rooms vs Price', 'Bathroom vs Price', 'Landsize vs Price', 'Lattitude vs Price', 'Longtitude vs Price']

importances = []

for i, ax in enumerate(axs.flat):
    if i < len(features):
        ax.scatter(df[features[i]], df['Price'])
        ax.set_xlabel(features[i])
        ax.set_ylabel('Price')
        ax.set_title(titles[i])

        correlation = df[features[i]].corr(df['Price'])
        importances.append(abs(correlation))

max_index = importances.index(max(importances))
most_affecting_feature = features[max_index]

if ax is axs[-1, -1]:
    ax.text(0.5,0.5, 'Most affecting feature is: ' + most_affecting_feature,
            horizontalalignment='center', verticalalignment='center',
            transform = ax.transAxes, fontsize=24)

plt.tight_layout()
```

<p align=center style="direction:rtl; fontsize:24px; color:#fff ;">
i've add importances and correlation to obtain most_affecting_feature and display it in last empty subplot for fun
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font size=4>
<ol dir=rtl start="5">
<li>
با استفاده از <b>5</b> ویژگی نام‌برده‌شده در مرحله‌ی قبل، مدلی طراحی کنید که قیمت خانه‌های ملبورن را پیش‌بینی کند.
</li>
</ol>
</font>

```python
# Do it.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error

features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']

X = df[features]
y = df['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
print('Model MAE:', mae)
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font size=4>
<ol dir=rtl start="6">
<li>
آیا مدل شما، از دقت خوبی برای پیش‌بینی قیمت خانه‌های ملبورن برخوردار است؟ درباره‌ی آن توضیح دهید. (امتیازی)
</li>
</ol>
</font>
</p>

<p align=center style="direction:rtl; color:white; font-weight:bolder">
با این وضع تراکم قیمت ها با فیچر هایی مثل تعداد اتاق های هتل و دقت 55 درصدی مدل خطی همه چیز واضحه
</p>

```python
plt.scatter(y_test, y_pred, color = "blue")

slope, intercept = np.polyfit(y_test, y_pred, 1)

abline_values = [slope * i + intercept for i in y_test]

plt.plot(y_test, abline_values, color = "red")

plt.title('Linear Regression Line vs Real Price')
plt.xlabel('Real Price')
plt.ylabel('Predicted Price')

plt.show()
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font size=4>
<ol dir=rtl start="7">
<li>
با استفاده از مدلی که طراحی کرده‌اید، قیمت خانه‌ای دارای <b>4</b> اتاق و <b>2</b> حمام و با اندازه زمین <b>500</b> که در عرض جغرافیایی <b>38.5-</b> و طول جغرافیایی <b>145</b> قرار دارد را تخمین بزنید.
</li>
</ol>

```python
# Do it.
# Estimate the price of a specific house
[[Rooms,Bathroom,Landsize,Lattitude,Longtitude]] = [[4, 2, 500, -38.5, 145]]
predicted_price = model.predict([[Rooms,Bathroom,Landsize,Lattitude,Longtitude]])
print('Estimated Price: $' + '{:,.2f}'.format(predicted_price[0]))

```
